```python
import bpy

# 1. Path to your addon zip file
addon_zip_path = "./retarget_bvh-4.2.1.zip"

# 2. Install the addon
bpy.ops.preferences.addon_install(filepath=addon_zip_path, overwrite=True)

# 3. Enable the addon
bpy.ops.preferences.addon_enable(module="retarget_bvh")

# 4. (Optional) Save preferences so addon stays active after Blender restarts
bpy.ops.wm.save_userpref()
```

```python
import bpy

# Print if retarget_bvh addon is active
print("retarget_bvh" in bpy.context.preferences.addons)

bvh_path = "bvh.bvh"
avatar = "avatar.glb"
bvh_name="bvh"
# Import BVH
bpy.ops.import_anim.bvh(filepath=bvh_path)

# Import GLB (glTF)
bpy.ops.import_scene.gltf(filepath=avatar)


bpy.ops.object.select_all(action='DESELECT')
bpy.data.objects[bvh_name].select_set(True)  # Source
bpy.context.view_layer.objects.active = bpy.data.objects["Armature"]  # Target

# Perform retargeting
result = bpy.ops.mcp.retarget_selected_to_active('EXEC_DEFAULT')
print(result)

# Select the armature and its children (meshes)
bpy.ops.object.select_all(action='DESELECT')
armature = bpy.data.objects["Armature"]
armature.select_set(True)
bpy.context.view_layer.objects.active = armature

# Select all child meshes
for child in armature.children:
    child.select_set(True)

# Export the selected armature and meshes to a .glb file
export_path = "retargeted_avatar.glb"
bpy.ops.export_scene.gltf(
    filepath=export_path,
    export_format='GLB',
    use_selection=True,
    export_apply=True
)
print(f"Exported to {export_path}")

```
