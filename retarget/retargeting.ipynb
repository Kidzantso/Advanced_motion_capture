```python
import bpy

# Print if retarget_bvh addon is active
print("retarget_bvh" in bpy.context.preferences.addons)

bvh_path = "bvh.bvh"
avatar = "avatar.glb"
bvh_name="bvh"
# Import BVH
bpy.ops.import_anim.bvh(filepath=bvh_path)

# Import GLB (glTF)
bpy.ops.import_scene.gltf(filepath=avatar)


bpy.ops.object.select_all(action='DESELECT')
bpy.data.objects[bvh_name].select_set(True)  # Source
bpy.context.view_layer.objects.active = bpy.data.objects["Armature"]  # Target

# Perform retargeting
result = bpy.ops.mcp.retarget_selected_to_active('EXEC_DEFAULT')
print(result)

def export_retargeted_avatar(output_path="retargeted_avatar.glb"):
    # Deselect all objects first
    bpy.ops.object.select_all(action='DESELECT')
    
    # Select the armature
    armature = bpy.data.objects["Armature"]
    armature.select_set(True)
    
    # Select the Wolf3D mesh
    for obj in bpy.data.objects:
        if obj.type == 'MESH' and obj.name.startswith('Wolf3D_Avatar'):
            obj.select_set(True)
    
    # Set the armature as active object
    bpy.context.view_layer.objects.active = armature
    
    # Export as GLB
    bpy.ops.export_scene.gltf(
        filepath=output_path,
        export_format='GLB',
        use_selection=True,
        export_animations=True
    )
    
    print(f"Avatar exported successfully to {output_path}")

# Call the function to export
export_retargeted_avatar()
```
